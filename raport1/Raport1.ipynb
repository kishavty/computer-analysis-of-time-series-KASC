```python
from scipy import stats
import numpy as np

def przedzialy_ufnosci(x, y, alpha):
    n = len(x)
    x_mean = np.mean(x)

    beta_1 = np.sum((x - x_mean) * (y - np.mean(y))) / np.sum((x - x_mean) ** 2)
    beta_0 = np.mean(y) - beta_1 * x_mean

    y_pred = beta_0 + beta_1 * x
    bledy = y - y_pred
    s_kwadrat = np.sum(bledy ** 2) / (n - 2)
    s = np.sqrt(s_kwadrat)
    se_beta_0 = s * np.sqrt(1 / n + x_mean * 2 / np.sum((x - x_mean) * 2))
    se_beta_1 = s / np.sqrt(np.sum((x - x_mean) ** 2))

    wyniki = {}
    z_value = stats.norm.ppf(1 - alpha / 2)

    t_value = stats.t.ppf(1 - alpha / 2, df=n - 2)
    ci_beta_0_nieznana_sigma = (beta_0 - t_value * se_beta_0, beta_0 + t_value * se_beta_0)
    ci_beta_1_nieznana_sigma = (beta_1 - t_value * se_beta_1, beta_1 + t_value * se_beta_1)

    return ci_beta_0_nieznana_sigma, ci_beta_1_nieznana_sigma, s_kwadrat


wyniki = przedzialy_ufnosci(X, Y, 0.05)
print(wyniki)
```

```python
def varr():
    varlist = []
    for i in range(2, len(e)):
        dosumy = (e[i])**2
        varlicz = np.sum(dosumy) / (i - 1)
        varlist.append(varlicz)
    return varlist

varlist = varr()
plt.plot(np.linspace(0, len(varlist), len(varlist)), varlist, color='black', ls="-", alpha=0.75, label="Wariancja residuów")
plt.legend()
plt.grid()
plt.title("Wariancja")
plt.show()
```

```python
def autocorrelation_emp(sigm, n, theta, hs):
    Z = np.random.normal(0, sigm**2, n)
    X = np.empty(n)
    ro_est = []

    for t in range(1, n):
        X[t] = Z[t] + theta * Z[t - 1]

    for h in hs:
        dosumy = np.empty(n - abs(h))
        for t in range(0, n - abs(h)):
            dosumy[t] = (X[t] - np.mean(X)) * (X[t + abs(h)] - np.mean(X))

        gamma = np.sum(dosumy) / (n - abs(h))

        dosumy_2 = (X - np.mean(X))**2

        ro_est.append(gamma / (np.sum(dosumy_2) / n))


        suma2 = 0
        for i in range(n):
            suma2 += (X[i]-np.mean(X))**2

    return ro_est



hs = np.arange(0, len(e))
n = len(e)
theta = 0.05
s = np.sqrt(sum((Y - esty)**2) / (len(e)) - 2)

result_emp = autocorrelation_emp(s, n, theta, hs)


plt.scatter(hs, result_emp, label = "empirical")
plt.title('Autokowariancja')
plt.grid(True)
plt.show()
```
